```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_table("../Datasets/breast-cancer-wisconsin.data", 
     sep=",", 
     names=['C1','C2','C3','C4','C5','C6','C7','C8','C9','C10','C11' ], 
     index_col="C1")
```

```python
df.describe()
```

```python
df.info()
```

```python
has_invalid=df[df['C7']=='?']
df.drop(list(has_invalid.index), axis=0, inplace=True)

df['C7']=pd.to_numeric(df['C7'])
```

```python
is2 = df[df['C11']==2] 
is4 = df[df['C11']==4]
```

```python
for i in range(2, 11):
    x2= is2['C'+str(i)].hist(bins=5, color='green', alpha=0.4, figsize=[10,5])
    x4= is4['C'+str(i)].hist(bins=5, color='red', alpha=0.4, figsize=[10,5]) 
    plt.title('C'+str(i))
    plt.show()
```
